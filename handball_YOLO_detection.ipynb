```python
from ultralytics import YOLO


model = YOLO("yolo12l.pt")

model.train(
    data="/content/Full-Handball-Dataset-1/data.yaml",
    epochs=100,
    batch=12,
    imgsz=800,
    device="cuda",
    workers=2,
    cache="disk",
    patience=20,
    project="/content/drive/MyDrive/handball_project2"
    name="handball_model"
)
```
